```python
import cv2
import numpy as np

def on_mouse(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        points.append((x, y))
        cv2.circle(img, (x, y), 5, (0, 255, 0), -1)
        cv2.imshow("Select Four Corners", img)

# Load image using a dialogue box
file_path = input("Enter the path of the image: ")
img = cv2.imread(file_path)

# Create a window and set the callback function
cv2.namedWindow("Select Four Corners")
cv2.setMouseCallback("Select Four Corners", on_mouse)

# Display the image
cv2.imshow("Select Four Corners", img)

# Initialize points
points = []

# Wait for the user to click 4 points
while len(points) < 4:
    cv2.waitKey(1)

# Convert the clicked points to numpy array
corners = np.array(points, dtype=np.int32)

# Close the window after 4 points are selected
cv2.destroyAllWindows()

# Display the original image with selected corners
img_with_corners = img.copy()
for pt in corners:
    cv2.circle(img_with_corners, tuple(pt), 5, (0, 0, 255), -1)

# Output coordinates
print("Top left corner coordinates:", corners[0])
print("Top right corner coordinates:", corners[1])
print("Bottom left corner coordinates:", corners[2])
print("Bottom right corner coordinates:", corners[3])

# Define the destination image size
w, h = 600, 300

# Define the source and destination points
pts1 = np.float32(corners)
pts2 = np.float32([[0, 0], [w, 0], [0, h], [w, h]])

# Calculate the perspective transformation matrix
matrix = cv2.getPerspectiveTransform(pts1, pts2)

# Apply perspective correction
op = cv2.warpPerspective(img, matrix, (w, h))

# Draw filled circles at the specified coordinates on the original image
for pt in pts1:
    cv2.circle(img, (int(pt[0]), int(pt[1])), 5, (0, 0, 255), cv2.FILLED)

# Display the original and unwarped images
cv2.imshow("Original", img)
cv2.imshow("Op", op)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
